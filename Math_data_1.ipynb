```python
urls = ["https://cdn.britannica.com/45/5645-050-B9EC0205/head-treasure-flower-disk-flowers-inflorescence-ray.jpg" , 
        "https://hips.hearstapps.com/hmg-prod.s3.amazonaws.com/images/pale-purple-cosmos-flower-1656258166.jpeg" ,
        "https://images.unsplash.com/photo-1604085572504-a392ddf0d86a?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxzZWFyY2h8Mnx8b3JhbmdlJTIwZmxvd2VyfGVufDB8fDB8fA%3D%3D&w=1000&q=80" ,
        "https://img.freepik.com/free-photo/purple-osteospermum-daisy-flower_1373-16.jpg?w=2000" , 
        "https://ichef.bbci.co.uk/news/976/cpsprodpb/957C/production/_111686283_pic1.png" ]
```

```python
images = [] #empty list to save images

for link in urls: #loop ovewr list of image urls
  print(link)

  im = io.imread(link) #load image from URL and store as numpy matrix (h, w, c)
  im = resize(im,(512,512)) #resize all images to same size 512 pixels wide by 512 pixels tall
  plt.imshow(im) #plot one image at a time
  plt.show() #draw image plot before next loop 

  images.append(im) #append adds new image called im to list of called images
```
